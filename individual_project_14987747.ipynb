## 3. Exploratory Data Analysis and Visualization

The parts for loading data into R and computing the mean values are both done in first part, which is data description.

### 3.1 Players Data

#### 3.1.1 Wrangling Data

```r
wrangling_players_data <- players_data |>
    filter(played_hours != 0, gender != "Prefer not to say") |>
    select(experience, subscribe, gender)

wrangling_players_data
```

#### 3.1.2 Visualization

```r
players_data_experience_plot <- wrangling_players_data |>
    ggplot(aes(x = experience,
               fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(x = "Players' Experience",
         y = "Percentage of Whether Players Subscribe",
         fill = "Whether Players Subscribe") +
    ggtitle("Relationship Between Players'Experience and Whether Players Subscribe")

players_data_gender_plot <- wrangling_players_data |>
    ggplot(aes(x = gender,
               fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(x = "Gender",
         y = "Percentage of Whether Players Subscribe",
         fill = "Whether Players Subscribe") +
    ggtitle("Relationship Between Gender and Whether Players Subscribe")

players_data_experience_plot
players_data_gender_plot
```

According to these two plots, it is obvious that players who are agender, other, and two-spirited, all subscribe.

### 3.2 Sessions Data

#### 3.2.1 Wrangling Data

```r
wrangling_sessions_data <- sessions_data |>
    select(original_start_time, original_end_time)

wrangling_sessions_data
```

#### 3.2.2 Visualization

```r
sessions_data_plot <- wrangling_sessions_data |>
    ggplot(aes(x = original_start_time,
               y = original_end_time)) +
    geom_point() +
    labs(x = "Original Start Time",
         y = "Original End Time") +
    ggtitle("Relationship Between Original Start Time and Original End Time")

sessions_data_plot
```

The relationship between the original start time and original end time is strong positive relationship.
